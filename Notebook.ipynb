```python
%matplotlib inline

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from counting_mnist.create_dataset import generate_sample
```

```python
mnist = tf.contrib.learn.datasets.mnist.read_data_sets('data/mnist/', reshape=False)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))
for ax in axes.ravel():
    clutter, density, count = generate_sample(mnist.train, 100, 28, 5, 8)
    ax.imshow(np.squeeze(clutter), cmap='gray')
    ax.imshow(np.squeeze(density), cmap='viridis', alpha=0.5)
    ax.axis('off')
    ax.set_title('{} even digits'.format(count))
plt.tight_layout()
```
